```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
page=requests.get("https://en.wikipedia.org/wiki/List_of_banks_in_India")
```

```python
web_data=BeautifulSoup(page.text,"html")   ##html to text
```

```python
table=web_data.find_all("table")[1]        ## search and select table
```

```python
column_names=table.find_all("th")         ##search and select th
```

```python
all_names_of_columns=[names.text.strip() for names in column_names]  ##select only text
```

```python
df=pd.DataFrame(columns=all_names_of_columns)
```

```python
df
```

```python
column_data=table.find_all("tr")
```

```python
for data in column_data[3:]:
    rows_data=data.find_all("td")
    individual_rows_data=[data.text.strip() for data in rows_data]
    print(individual_rows_data)
    length=len(df)
    df.loc[length]=individual_rows_data
```
