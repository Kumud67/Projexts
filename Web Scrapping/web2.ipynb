```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get("https://en.wikipedia.org/wiki/List_of_Indian_IT_companies")
```

```python
soup=BeautifulSoup(page.text,"html")
```

```python
print(soup)
```

```python
table=soup.find_all("table")[2]
```

```python
print(table)
```

```python
col_names=table.find_all("th")
```

```python
col_table_names=[names.text.strip() for names in col_names]
    
```

```python
print(col_table_names)
```

```python
import pandas as pd
```

```python

df= pd.DataFrame(columns=col_table_names)
df
```

```python
column_data=table.find_all("tr")
```

```python
for rows in column_data[1:]:
    rows_data=rows.find_all("td")
    individual_rows_data=[data.text.strip() for data in rows_data]
    print(individual_rows_data)
    length=len(df)
    df.loc[length]=individual_rows_data
```
